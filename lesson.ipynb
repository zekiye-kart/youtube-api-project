```python
api_service_name = "youtube"
api_version = "v3"

channel_ids = [
    "UCYO_jab_esuFRV4b17AJtAw",
    # Can add more channels
]

# Get credentials and create an API client
youtube = build(api_service_name, api_version, developerKey=YOUTUBE_API_KEY)
```

```python
def get_channel_stats(youtube, channel_ids):
    all_data = []

    request = youtube.channels().list(
        part="snippet,contentDetails,statistics",
        id=",".join(channel_ids)
    )

    response = request.execute()

    # loop through items in response
    for item in response["items"]:
        data = {'channelName': item['snippet']['title'],
                'subscribers': item['statistics']['subscriberCount'],
                'views': item['statistics']['viewCount'],
                'totalVideos': item['statistics']['videoCount'],
                'playlistId': item['contentDetails']['relatedPlaylists']['uploads']
        }
        all_data.append(data)
    
    return pd.DataFrame(all_data)
```

```python
channel_ids = get_channel_stats(youtube, channel_ids)
```

Bu fonksiyon, verilen playlist_id (oynatma listesi ID'si) ile belirtilen oynatma listesindeki videoların ID'lerini çeker. 

Fonksiyon, bu video ID'lerini içeren bir liste döndürür. Bu listenin kullanımı, elde edilen video ID'lerini başka bir amaç için kullanmanıza bağlıdır, örneğin bu videoların detaylarını almak veya başka bir işlem gerçekleştirmek.

```python
def get_video_ids(youtube, playlist_id):
    video_ids = []

    request = youtube.playlistItems().list(
            part="snippet,contentDetails",
            playlistId=playlist_id
        )
    response = request.execute()

    for item in response['items']:
        video_ids.append(item['contentDetails']['videoId'])
        
    return video_ids
```

```python
playlist_id = "UUYO_jab_esuFRV4b17AJtAw"

video_ids = get_video_ids(youtube, playlist_id)
```

Bir YouTube oynatma listesine ait video ID'lerini almak için sayfa token'ları kullanarak birden fazla sayfa üzerinde gezinmeyi destekler. Bu, bir oynatma listesindeki tüm videoları çekmek için kullanışlıdır, çünkü bir oynatma listesi çok sayıda video içerebilir ve YouTube API'ları genellikle sayfalandırılmış sonuçlar döndürür.

- Her bir sayfa, maxResults parametresi ile belirtilen kadar (bu örnekte 50) video içerir ve pageToken parametresi ile bir sonraki sayfa belirlenir. Bu işlem, API'nin döndürdüğü her sayfa için tekrarlanır, ta ki nextPageToken None olana kadar. Bu sayede tüm oynatma listesi üzerinde dolaşılır ve video ID'leri toplanır.

```python
def get_video_ids(youtube, playlist_id):
    video_ids = []

# İlk isteği oluşturuyoruz, maxResults parametresi ile her istekte en fazla 50 video çekilecek
    request = youtube.playlistItems().list(
            part="snippet,contentDetails",
            playlistId=playlist_id,
            maxResults=50
        )
    response = request.execute()
    
# İlk sayfadaki her bir öğe (item) için video ID'yi çekiyoruz ve listeye ekliyoruz
    for item in response['items']:
        video_ids.append(item['contentDetails']['videoId'])

# İlk sayfa token'ını alıyoruz
    next_page_token = response.get('nextPageToken')
    
    
# Daha fazla sayfa varsa, döngü içinde gezinmeye devam ediyoruz
    while next_page_token is not None:
        # Bir sonraki sayfa için isteği oluşturuyoruz, pageToken parametresi ile bir sonraki sayfanın token'ını belirtiyoruz
        request = youtube.playlistItems().list(
                part="snippet,contentDetails",
                playlistId=playlist_id,
                maxResults=50,
                pageToken=next_page_token  # Bir sonraki sayfanın token'ı
            )
        response = request.execute()

         # Her bir öğe için video ID'yi çekiyoruz ve listeye ekliyoruz
        for item in response['items']:
            video_ids.append(item['contentDetails']['videoId'])

          # Bir sonraki sayfa token'ını alıyoruz, eğer yoksa döngü sonlanır
        next_page_token = response.get('nextPageToken')    

        
    return video_ids
```

YouTube API tarafından kullanılan "pageToken" veya "nextPageToken" kavramları, sayfalandırılmış sonuçların yönetilmesine yardımcı olan belirli bir işarettir. Sayfa token'ları, API'nin döndürdüğü veri setini bölümlere ayırmak ve kullanıcılara belirli bir sayfa veya belirli bir konumdan sonraki verileri almak için kullanılır.

```python
playlist_id = "UUYO_jab_esuFRV4b17AJtAw"

video_ids = get_video_ids(youtube, playlist_id)
```
